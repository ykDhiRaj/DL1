```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['ChestPainType', 'RestingECG','ST_Slope'], drop_first=True)
```

```python
df.head()
```

```python
categorical_cols = ["ChestPainType_ATA", "ChestPainType_NAP", "ChestPainType_TA","RestingECG_Normal","RestingECG_ST","ST_Slope_Flat","ST_Slope_Up"]
df[categorical_cols] = df[categorical_cols].apply(le.fit_transform)
```

```python
df.head()
```

```python
X = df.drop('HeartDisease', axis=1)
y = df['HeartDisease']
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=32,
    verbose=1
)
```
